```python
#check lagged stock_ordered vs stock_received
def lag_feature( df,lags, cols ):
    for col in cols:
        print(col)
        tmp = df[[ "site_code", "product_code","date_block_num",col ]]
        for i in lags:
            shifted = tmp.copy()
            shifted.columns = ["site_code", "product_code","date_block_num", col + "_lag_"+str(i)]
            shifted['date_block_num'] = shifted['date_block_num'] + i
            df = pd.merge(df, shifted, on=['date_block_num','site_code','product_code'], how='left')
    return df
```

```python
train3 = lag_feature(train3, [1], ['stock_ordered'])
```
